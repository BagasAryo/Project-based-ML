## 1. Pie Chart
---
**Pie Chart** digunakan untuk menunjukkan seberapa banyak dari setiap jenis kategori dalam dataset berbanding dengan keseluruhan.

```python
# visualisasi pie chart 
flavors = ('Chocolate', 'Vanilla', 'Pistachio', 'Mango', 'Strawberry')
votes = (12, 11, 4, 8, 7)
plt.pie(
    votes,
    labels=flavors,
)
plt.show()
```

Melihat berapa persen kontribusi masing-masing rasa es krim, kita bisa menggunakan argumen **autopact**

```python
# melihat persen kontribusi pada pie chart
flavors = ('Chocolate', 'Vanilla', 'Pistachio', 'Mango', 'Strawberry')
votes = (12,11,4,8,7)
plt.pie(
    votes,
    labels=flavors,
    autopct='%1.1f%%',
)
plt.show()
```

```python
# kostumisasi warna pie chart
flavors = ('Chocolate', 'Vanilla', 'Pistachio', 'Mango', 'Strawberry')
votes = (12,11,4,8,7)
colors = ('#FF76CE', '#FDFFC2', '#94FFD8', '#A3D8FF', '#7EA1FF')
plt.pie(
    votes,
    labels=flavors,
    autopct='%1.1f%%',
    colors=colors,
)
plt.show()
```

Melakukan highlight terhadap terhadap data mangga, kemudian mendorong data mangga keluar sebesar 0,5 sambil menjaga semua potongan lainnya tetap berada ditengah bisa menggunakan argumen **explode**

```python
# menambah highlight terhadap data
flavors = ('Chocolate', 'Vanilla', 'Pistachio', 'Mango', 'Strawberry')
votes = (15, 10, 5, 9, 11)
colors = ('#FF76CE', '#FDFFC2', '#94FFD8', '#A3D8FF', '#7EA1FF')
explode = (0, 0, 0, 0.15, 0)
plt.pie(
    votes,
    labels=flavors,
    autopct='%1.1f%%',
    colors=colors,
    explode=explode,
)
plt.show()
```

Menambahkan sedikit efek tiga dimensi, bayangan dengan argumen **shadow**

```python
# menambah bayangan visualisasi
flavors = ('Chocolate', 'Vanilla', 'Pistachio', 'Mango', 'Strawberry')
votes = (17, 12, 5, 4, 11)
colors = ('#FF76CE', '#FDFFC2', '#94FFD8', '#A3D8FF', '#7EA1FF')
explode = (0, 0, 0, 0.15, 0)
plt.pie(
    votes,
    labels=flavors,
    autopct='%1.1f%%',
    colors=colors,
    explode=explode,
    shadow=True
)
plt.show()
```

Menambahkan judul menggunakan **plt.title()**

```python
# menambahkan judul
flavors = ('Chocolate', 'Vanilla', 'Pistachio', 'Mango', 'Strawberry')
votes = (13, 8, 5, 4, 11)
colors = ('#FF76CE', '#FDFFC2', '#94FFD8', '#A3D8FF', '#7EA1FF')
explode = (0, 0, 0, 0.15, 0)

plt.title('Favorite Ice Cream Flavors')
plt.pie(
    votes,
    labels=flavors,
    autopct='%1.1f%%',
    colors=colors,
    explode=explode,
    shadow=True
)
plt.show()
```

## 2. Correlation and Causation
---
**Korelasi** merupakan suatu pengukuran sejauh mana nilai saling ketergantungan antar variabel. **Causation** merupakan hubungan antara sebab dan akibat antara dua variable penting untuk mengetahui perbedaan antara keduanya dan bahwa korelasi tidak mendeskripsikan sebab-akibat. Menentukan korelasi jauh lebih sederhana sebab memerlukan analisis lebih lanjut.

### Korelasi Pearson
Korelasi Pearson menguku rketergantungan linier antara dua variabel X dan Y. Koefisien yang dihasilkan adalah nilai antara -1 dan 1 inklusif, di mana : 
* 1: Total korelasi linier positif.
* 0: Tidak ada korelai linier, kedua variabel kemungkinan besar tidak saling mempengaruhi.
* -1: Total korelasi linier negatif. Pearson Correlation adalah motde default dari fungsi **"corr**. 

Seperti sebelumnya kita dapat menghitung Korelasi Pearson dari variabel 'int64' atau 'float64'. Terkadang kita ingin mengetahui signifikasi dari estimasi korelasi, kita dapat menggunakan p-value. P-value: 
Berapa nilai P ini? Nilai P adalah nilai probabilitas bahwa korelasi antara kedua variabel ini signifikan secara statistik. Biasanya, kita memilih tingkat signifikansi 0,05 yang berarti bahwa kami yakin 95% korelasi antar variabel signifikan. 
Dengan konvensi, ketika :
* nilai nilai p adalah $<$ 0,001: kami katakan ada bukti kuat bahwa korelasinya signifikan.
* nilai p adalah $<$ 0,05: terdapat bukti moderat bahwa korelasi tersebut signifikan.
* nilai p adalah $<$ 0,1: ada bukti lemah bahwa korelasinya signifikan.
* nilai p adalah $>$ 0,1: tidak ada bukti bahwa korelasi tersebut signifikan.
Kita dapat menggunakan library scipy untuk menghitung korelasi dan p-value


```python
# import library scipy
from scipy import stats
```

```python
# menghitung koefisien korelasi pearson dari 'wheels-base' & 'price'
pearson_coef, p_value = stats.pearsonr(df['wheel-base'], df['price'])
print('The Pearson Correlation Coefficienet is', pearson_coef, ' with a P-value of P =', p_value)
```
